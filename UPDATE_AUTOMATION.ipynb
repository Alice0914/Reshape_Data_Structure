```python
import pandas as pd
import numpy as np
```

```python
s_main_df = pd.read_excel('0778_MAIN_0811.xlsx')
```

```python
s_main_df.head()
```

```python
s_p07 = pd.read_excel('0778_P07_0811.xlsx')
```

```python
s_p07.head()
```

```python
s_p40 = pd.read_excel('0778_P40_0811.xlsx')
```

```python
s_p40.head()
```

```python
p07_pivot = pd.pivot_table(s_p07, index = 'P08_NUM', columns = ['PF_TYPE', 'SALES_ORG', 'VTWEG', 'SPART', 'PARZA'], values = 'PF_NUM')
p07_pivot
```

```python
p40_pivot = s_p40.pivot_table(index = 'P40_BP', columns = ['PF_TYPE', 'SALES_ORG', 'VTWEG', 'SPART', 'PARZA'], values = 'PF_NUM', aggfunc='first')
p40_pivot
```

```python
total_pivot_p07 = s_main_df.merge(p07_pivot, left_on = 'P08 NUMBER', right_index = True, how = 'left')
total_pivot_p07
```

```python
total_pivot_p07_p40 = total_pivot_p07.merge(p40_pivot, left_on = 'BP Number', right_index = True, how='left')
total_pivot_p07_p40
```

```python
total_pivot_p07_p40.to_excel('Main_Pivot_0812_2.xlsx')
```
